```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
import matplotlib.pyplot as plt
```

```python
size = 64
background = Image.new('RGBA', (size, size), (0, 0, 0, 0))
size_ = 8
for i, pixel in enumerate(background.getdata()):
    i, j = i % size, i // size
    if (i // size_ + j // size_) % 2 == 0:
        background.putpixel((i, j), (127, 127, 127, 255))
    else:
        background.putpixel((i, j), (64, 64, 64, 255))
```

```python
font = ImageFont.truetype('DejaVuSans-Bold.ttf', 64)
images = []
offset = (10, -4)
for i in range(1, 10):
    canvas = Image.new('RGBA', (size, size), (0, 0, 0, 0))
    draw = ImageDraw.Draw(canvas)
    draw.text(offset, f'{i}', font=font, fill=(255, 255, 255, 255))
    canvas.save(f'../graphics/icons/pointer-{i}.png')
    images.append(canvas.copy())

fig, axs = plt.subplots(3, 3)
for i, ax in enumerate(axs.flat):
    ax.imshow(background)
    ax.imshow(images[i])
    ax.axis('off')
```
